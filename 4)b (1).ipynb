```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

df = pd.DataFrame(data)

print(df.head())

df['TotalValue'] = df['Price'] * df['Quantity']
print(df[['Product','TotalValue']])

print(df['Price'].mean())
print(df['Quantity'].max())

df_sorted = df.sort_values(by='TotalValue', ascending=False)
print(df_sorted[['Product','TotalValue']])

plt.bar(df['Product'], df['TotalValue'], color='skyblue')
plt.ylabel('Total Value')
plt.title('Inventory Total Value per Product')
plt.show()
```
