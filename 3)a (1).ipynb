```python
import numpy as np

f = 50

R = 100
C = 100e-6
L = 0.5

Z_R = R
Z_C = -1j / (2 * np.pi * f * C)
Z_L = 1j * 2 * np.pi * f * L

circuit_type = "series"

if circuit_type.lower() == "series":
    Z_total = Z_R + Z_C + Z_L
elif circuit_type.lower() == "parallel":
    Z_total = 1 / (1/Z_R + 1/Z_C + 1/Z_L)

print("Total Impedance:", Z_total)
print("Magnitude:", abs(Z_total))
print("Phase (radians):", np.angle(Z_total))
```
