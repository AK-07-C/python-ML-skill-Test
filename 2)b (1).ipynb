```python
import numpy as np

grades = np.array([
[78,82,80],[65,70,68],[90,88,92],[55,60,58],[72,75,70],
[85,87,83],[60,65,62],[95,92,94],[70,68,72],[88,85,90],
[76,80,78],[69,72,70],[84,86,82],[58,60,55],[91,89,93],
[73,75,77],[66,68,64],[79,82,81],[87,90,88],[62,65,63],
[74,78,76],[81,83,80],[68,70,69],[92,94,91],[77,79,75],
[83,85,84],[59,61,60],[71,73,72],[86,88,87],[64,66,65]
])

avg = np.mean(grades, axis=0)
high = np.max(grades, axis=0)
low = np.min(grades, axis=0)

print("Average marks:", avg)
print("Highest marks:", high)
print("Lowest marks:", low)

total = np.sum(grades, axis=1)
top = np.argmax(total)

print("\nTop performing student:", top + 1)
print("Total marks:", total[top])
```
