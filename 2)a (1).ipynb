```python
import numpy as np

steps = np.array([
5000, 6200, 7100, 8000, 7500, 9000, 10000,
6500, 7200, 8800, 9200, 11000, 10500, 9800,
7600, 8300, 8900, 9400, 10200, 9700, 8800,
7900, 8100, 8600, 9100, 9500, 9900, 10400,
10800, 11200
])

total = np.sum(steps)
avg = np.mean(steps)
high = np.max(steps)
low = np.min(steps)

print("Total steps:", total)
print("Average steps per day:", avg)
print("Maximum steps:", high)
print("Minimum steps:", low)

threshold = 8000
print("\nHigh activity days:")
for i in range(len(steps)):
    if steps[i] > threshold:
        print("Day", i+1, ":", steps[i])

print("\nTrend analysis:")
for i in range(1, len(steps)):
    if steps[i] > steps[i-1]:
        print("Day", i, "to Day", i+1, ": Increase")
    elif steps[i] < steps[i-1]:
        print("Day", i, "to Day", i+1, ": Decrease")
    else:
        print("Day", i, "to Day", i+1, ": No change")
```
