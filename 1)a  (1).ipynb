```python
import numpy as np

marks = [78,45,89,56,92,61,47,73,84,39,68,90,55,49,77,81,64,52,58,71]

data = np.array(marks)

avg = np.mean(data)
high = np.max(data)
low = np.min(data)

print("Average:", avg)
print("Highest:", high)
print("Lowest:", low)

print("\nPass/Fail")
for i in range(len(data)):
    if data[i] >= 50:
        print("Student", i+1, "Pass")
    else:
        print("Student", i+1, "Fail")

print("\nGrades")
for i in range(len(data)):
    m = data[i]
    
    if m >= 85:
        g = "A"
    elif m >= 70:
        g = "B"
    elif m >= 60:
        g = "C"
    elif m >= 50:
        g = "D"
    else:
        g = "F"
        
    print("Student", i+1, "Grade:", g)
```
