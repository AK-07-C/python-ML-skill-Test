```python
import numpy as np

stock1 = np.array([
    [100, 105, 98],
    [102, 107, 99],
    [101, 106, 97],
    [103, 108, 100],
    [104, 109, 101]
])

stock2 = np.array([
    [98, 103, 95],
    [100, 105, 97],
    [99, 104, 96],
    [101, 106, 98],
    [102, 107, 99]
])

sum_matrix = stock1 + stock2
diff_matrix = stock1 - stock2

print("Element-wise Sum of Stock Prices:\n", sum_matrix)
print("\nElement-wise Difference of Stock Prices:\n", diff_matrix)
```
